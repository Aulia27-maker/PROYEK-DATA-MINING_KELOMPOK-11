```python
from imblearn.over_sampling import SMOTE
```

Kode tersebut digunakan untuk mengimpor SMOTE (Synthetic Minority Over-sampling Technique) dari library Imbalanced-Learn. SMOTE merupakan metode yang digunakan untuk mengatasi masalah ketidakseimbangan jumlah data antar kelas (imbalanced data) pada proses klasifikasi. Ketidakseimbangan data terjadi ketika jumlah data pada satu kelas jauh lebih banyak dibandingkan kelas lainnya, misalnya jumlah siswa yang berstatus Tuntas jauh lebih banyak daripada siswa yang berstatus Tidak Tuntas. Kondisi tersebut dapat menyebabkan model lebih cenderung memprediksi kelas mayoritas sehingga performa klasifikasi menjadi kurang optimal. SMOTE bekerja dengan membuat data sintetis baru pada kelas minoritas berdasarkan karakteristik data yang sudah ada, sehingga jumlah data pada setiap kelas menjadi lebih seimbang. Dalam penelitian ini, SMOTE digunakan sebelum proses pelatihan model klasifikasi agar algoritma dapat mempelajari kedua kelas secara lebih baik dan menghasilkan prediksi yang lebih akurat, adil, serta tidak bias terhadap kelas mayoritas.

```python
from imblearn.pipeline import Pipeline as ImbPipeline
```

Kode tersebut digunakan untuk mengimpor Pipeline dari library Imbalanced-Learn dan memberinya nama singkat (alias) ImbPipeline. Pipeline berfungsi untuk menggabungkan beberapa tahapan pengolahan data ke dalam satu alur kerja yang terintegrasi, seperti proses preprocessing data, penyeimbangan data menggunakan SMOTE, dan pelatihan model klasifikasi. Penggunaan ImbPipeline sangat penting ketika menerapkan SMOTE karena teknik ini harus dijalankan hanya pada data pelatihan (training data) agar tidak terjadi data leakage yang dapat menyebabkan hasil evaluasi model menjadi tidak valid. Dalam penelitian ini, ImbPipeline digunakan untuk memastikan bahwa setiap data yang diproses akan melalui tahapan yang sama secara berurutan, yaitu transformasi fitur menggunakan ColumnTransformer, penyeimbangan kelas menggunakan SMOTE, dan pelatihan model klasifikasi seperti Random Forest, Logistic Regression, atau algoritma lainnya. Dengan demikian, proses pengolahan data menjadi lebih sistematis, efisien, dan meminimalkan kesalahan dalam implementasi model machine learning.

```python
# =========================
# PENENTUAN TARGET BERDASARKAN KKM
# =========================
# KKM = 75
# Nilai 75 sampai 100 = Tuntas
# Nilai di bawah 75 = Tidak Tuntas
# CATATAN:
# Bagian ini hanya untuk target pelatihan model.
# Hasil Tuntas/Tidak Tuntas tidak ditampilkan dulu.

KKM = 75

data['KETUNTASAN'] = np.where(
    (data['NILAI'] >= KKM) & (data['NILAI'] <= 100),
    'Tuntas',
    'Tidak Tuntas'
)
```

Selanjutnya, fungsi np.where() digunakan untuk melakukan klasifikasi secara otomatis terhadap setiap data siswa. Secara sederhana, aturan yang diterapkan adalah:

**3. Preprocessing Fitur**

Fitur yang dipakai pada proses pemodelan diantaranya adalah gender, model pembelajaran, dan nilai. Ketiga fitur tersebut digunakan karena dapat merepresentasikan karakteristik murid, jenis model pembelajaran, dan nilai penilaian terhadap kemampuan berpikir kreatif matematika. Sebelum melakukan proses pemodelan, beberapa fitur dengan bentuk kategorikal perlu dilakukan transformasi misalnya, gender dan model pembelajaran yang telah ditransformasi menggunakan teknik *one hot encoder.* Variabel yang bersifat numerik, misalnya saja nilai, akan dinormalisasi menggunakan teknik *standard scaler*. Ini dibutuhkan mengingat beberapa metode dalam machine learning sangat sensitif terhadap bentuk dan skala data.

```python
# =========================
# PREPROCESSOR
# =========================

categorical_features = ['GENDER', 'MODEL_PEMBELAJARAN']
numeric_features = ['NILAI']

preprocessor = ColumnTransformer(
    transformers=[
        ('categorical', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('numeric', StandardScaler(), numeric_features)
    ]
)
```

Kode ini membuat preprocessor untuk mentransformasi fitur sebelum dimasukkan ke model machine learning. Terdapat dua jenis fitur: kategorikal ('GENDER' dan 'MODEL_PEMBELAJARAN') serta numerik ('NILAI'). Menggunakan ColumnTransformer, fitur kategorikal diubah dengan OneHotEncoder (mengubah kategori seperti Laki/Perempuan atau CBL/PBL menjadi kolom biner 0/1, dengan handle_unknown='ignore' agar aman jika ada kategori baru saat prediksi). Sementara itu, fitur numerik 'NILAI' dinormalisasi dengan StandardScaler (mengubahnya agar memiliki rata-rata 0 dan standar deviasi 1, sehingga semua fitur berada dalam skala yang sama). Preprocessor ini siap digunakan dengan metode .fit_transform() pada data latih dan .transform() pada data validasi atau uji.

Selain itu, *Gaussian Naïve* Bayes digunakan karena memiliki proses klasifikasi yang sederhana dan efisien terutama pada dataset dengan distribusi probabilistik.

Sebelum melakukan klasifikasi, dataset dibagi menjadi data *training* dan data *testing*. Data *training* digunakan untuk membangun model klasifikasi, sedangkan data *testing* berfungsi untuk mengukur seberapa efektif model dalam memprediksi data yang baru.Akhirnya, hasil klasifikasi dari tiap-tiap algoritma dibandingkan menggunakan sejumlah metrik pengukuran guna mengambil algoritma yang paling efektif mengklasifikasikan ketuntasan kemampuan berpikir kreatif murid di model pembelajaran CBL dan PBL.


```python
# =========================
# MODEL KLASIFIKASI
# =========================

models = {
    'LogisticRegression': LogisticRegression(
        max_iter=1000,
        class_weight='balanced',
        random_state=42
    ),

    'RandomForest': RandomForestClassifier(
        n_estimators=300,
        random_state=42,
        class_weight='balanced'
    ),

    'GradientBoosting': GradientBoostingClassifier(
        random_state=42
    ),

    'AdaBoost': AdaBoostClassifier(
        n_estimators=300,
        learning_rate=0.05,
        random_state=42
    ),
    'KNN': KNeighborsClassifier(
        n_neighbors=5
    ),

    'SVM': SVC(
        class_weight='balanced',
        random_state=42
    ),

    'GaussianNB': GaussianNB(),

    'XGBoost': XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=5,
        random_state=42,
        eval_metric='logloss'
    ),

    'LightGBM': LGBMClassifier(
        n_estimators=300,
        learning_rate=0.05,
        random_state=42,
        class_weight='balanced',
        verbose=-1
    )
}
```

Kode ini mendefinisikan 9 model klasifikasi yang akan dibandingkan kinerjanya untuk memprediksi ketuntasan siswa. Setiap model disimpan dalam dictionary models dengan konfigurasi spesifik:


*   LogisticRegression: Menggunakan max_iter=1000 agar konvergen, class_weight='balanced' untuk menangani ketidakseimbangan kelas.

*   RandomForest: 300 pohon keputusan (n_estimators=300) dengan penyeimbangan kelas.

*   GradientBoosting: Model boosting default dengan random_state=42 untuk reproduksibilitas.

*   AdaBoost: 300 estimator dengan learning_rate=0.05 (langkah update kecil untuk stabilitas).

*   KNN: Menggunakan 5 tetangga terdekat (n_neighbors=5).
*   SVM: Support Vector Machine dengan penyeimbangan kelas.


*   GaussianNB: Naive Bayes asumsi distribusi normal, tanpa parameter khusus.


*   XGBoost: 300 pohon, learning_rate=0.05, max_depth=5, dengan metrik evaluasi 'logloss'.


*   LightGBM: 300 pohon, learning rate kecil, kelas seimbang, dan verbose=-1 untuk menyembunyikan log.

Semua model yang mendukung random_state diset agar hasil konsisten. Kumpulan model ini akan dilatih dan dievaluasi untuk menemukan algoritma terbaik dalam mengklasifikasikan ketuntasan siswa berdasarkan fitur gender, model pembelajaran, dan nilai tes.

**5. Penanganan**

Ketidakseimbangan Kelas
Dengan mempertimbangkan penyebaran data, ada lebih banyak murid yang dikelompokkan sebagai kategori tuntas dibandingkan murid yang tidak tuntas. Hal ini berarti bahwa terdapat ketidakseimbangan antara kelas tersebut. Solusinya adalah dengan menggunakan SMOTE yang diterapkan dalam data training. SMOTE dapat membantu menyeimbangkan data kelas minoritas dengan pembuatan data sintetis, yang mana akan mencegah model belajar secara dominan mengenai data kelas mayoritas saja. Penggunaan SMOTE ini dilakukan dalam *pipeline* untuk memastikan bahwa penyeimbangan data hanya dilakukan pada data training.


```python
# =========================
# PIPELINE MODEL
# =========================

pipelines = {}

for model_name, model in models.items():
    pipelines[model_name] = ImbPipeline(steps=[
        ('preprocessor', preprocessor),
        ('smote', SMOTE(random_state=42, k_neighbors=3)),
        ('model', model)
    ])
```

Kode ini membangun pipeline untuk setiap model menggunakan ImbPipeline (pipeline khusus untuk data tidak seimbang dari library imbalanced-learn). Prosesnya terdiri dari tiga langkah berurutan:


*   Preprocessor (preprocessor): Mengubah fitur kategorikal dengan OneHotEncoder dan fitur numerik dengan StandardScaler.

*   SMOTE (SMOTE(random_state=42, k_neighbors=3)): Membangkitkan sampel sintetis untuk kelas minoritas (siswa tidak tuntas) agar data menjadi seimbang, menggunakan 3 tetangga terdekat. Parameter random_state=42 memastikan reproduksibilitas.
*   Model (model): Masing-masing dari 9 model klasifikasi yang sudah didefinisikan sebelumnya.

Untuk setiap model dalam dictionary models, loop akan membuat pipeline dengan nama yang sama dan menyimpannya ke dalam dictionary pipelines. Hasilnya, pipelines['RandomForest'] misalnya, berisi pipeline lengkap yang siap digunakan dengan metode .fit(X_train, y_train) untuk melatih model dan .predict(X_val, y_val) untuk memprediksi data validasi. Pendekatan pipeline ini memastikan semua transformasi data (encoding, scaling, SMOTE) dilakukan secara konsisten dan terhindar dari kebocoran data.

**6. Evaluasi Model**

Penilaian tahap evaluasi dilakukan untuk melihat tingkat performa setiap algoritma klasifikasi dalam mengenali keberhasilan pemahaman berpikir kreatif anak-anak. Evaluasi model penelitian ini menggunakan beberapa metrik pengujian, seperti: *Accuracy, Precision, Recall* dan *F1-Score*.

Di samping itu, studi ini juga memanfaatkan *confusion matrix* untuk mengevaluasi jumlah prediksi yang tepat dan yang tidak tepat yang diperoleh dari model klasifikasi. Penggunaan beberapa metrik evaluasi bertujuan agar hasil pengujian model dapat dianalisis secara lebih menyeluruh. Model dengan nilai evaluasi terbaik dipilih sebagai model yang paling optimal dalam mengklasifikasikan ketuntasan kemampuan berpikir kreatif murid pada model pembelajaran CBL dan PBL. Hasil evaluasi tersebut selanjutnya digunakan untuk menganalisis pola ketuntasan murid dan memahami berbagai unsur yang berpengaruh terhadap kemampuan berpikir kreatif murid.


```python
# =========================
# TRAINING DAN EVALUASI MODEL
# =========================

results = []

for model_name, pipeline in pipelines.items():
    print("Melatih model:", model_name)

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_val)

    accuracy = accuracy_score(y_val, y_pred)
    precision_macro = precision_score(y_val, y_pred, average='macro', zero_division=0)
    recall_macro = recall_score(y_val, y_pred, average='macro', zero_division=0)
    f1_macro = f1_score(y_val, y_pred, average='macro', zero_division=0)

    results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision_macro': precision_macro,
        'Recall_macro': recall_macro,
        'F1_macro': f1_macro
    })

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(
    by=['Accuracy', 'F1_macro'],
    ascending=False
).reset_index(drop=True)

(results_df)
```

Kode tersebut melakukan proses training dan evaluasi terhadap sembilan model machine learning yang telah didefinisikan sebelumnya dalam bentuk pipeline. Pertama, dibuat list kosong bernama results untuk menampung hasil evaluasi. Kemudian, dilakukan perulangan (loop) untuk setiap model dalam dictionary pipelines. Pada setiap iterasi, model dilatih menggunakan pipeline.fit(X_train, y_train) lalu melakukan prediksi pada data validasi dengan pipeline.predict(X_val). Setelah itu, dihitung empat metrik evaluasi yaitu accuracy (akurasi), precision macro (presisi makro), recall macro (recall makro), dan F1 macro menggunakan fungsi-fungsi dari scikit-learn. Parameter average='macro' digunakan agar perhitungan tidak terpengaruh oleh ketidakseimbangan kelas, sedangkan zero_division=0 mencegah pembagian dengan nol. Setiap hasil disimpan sebagai dictionary ke dalam list results. Terakhir, hasil diubah menjadi DataFrame menggunakan pd.DataFrame(results), diurutkan berdasarkan kolom 'Accuracy' dan 'F1_macro' secara menurun (descending), lalu ditampilkan.


Output yang dihasilkan menunjukkan hasil evaluasi dari sembilan model yang telah dilatih. Enam model yaitu RandomForest, GradientBoosting, AdaBoost, SVM, XGBoost, dan LightGBM mencapai akurasi sempurna sebesar 1,000000 atau 100%, dengan nilai precision macro, recall macro, dan F1 macro semuanya 1,0. Ini berarti keenam model tersebut mampu memprediksi seluruh data validasi dengan tepat tanpa kesalahan sedikit pun, baik untuk kelas tuntas maupun tidak tuntas. **8. Perbandingan Kelas CBL dan PBL**

Tahap terakhir adalah membandingkan hasil klasifikasi antara kelas CBL dan PBL. Perbandingan dilakukan dengan melihat jumlah murid yang tuntas dan tidak tuntas, persentase ketuntasan, serta rata-rata nilai pada masing-masing model pembelajaran. Model pembelajaran yang memiliki persentase ketuntasan lebih tinggi dan rata-rata nilai yang lebih baik dapat dianggap menghasilkan dampak yang lebih positif pada keterampilan berpikir kreatif murid. Oleh karena itu, pendekatan yang diterapkan dalam penelitian ini tidak hanya berfungsi untuk mengklasifikasikan ketuntasan murid, tetapi juga membantu membandingkan efektivitas pembelajaran CBL dan PBL berdasarkan hasil tes kemampuan berpikir kreatif matematika.


```python
# =========================
# SUMMARY MODEL PEMBELAJARAN
# =========================

summary_model = data.groupby('MODEL_PEMBELAJARAN').agg(
    Jumlah_Siswa=('NILAI', 'count'),
    Rata_Rata_Nilai=('NILAI', 'mean'),
    Prediksi_Tuntas=('PREDIKSI_MODEL_TERBAIK', lambda x: (x == 'Tuntas').sum()),
    Prediksi_Tidak_Tuntas=('PREDIKSI_MODEL_TERBAIK', lambda x: (x == 'Tidak Tuntas').sum())
).reset_index()

summary_model['Persentase_Prediksi_Tuntas'] = (
    summary_model['Prediksi_Tuntas'] / summary_model['Jumlah_Siswa'] * 100
)

print("\nSummary Model Pembelajaran:")
print(summary_model)


for model in summary_model['MODEL_PEMBELAJARAN']:
    data_model = summary_model[summary_model['MODEL_PEMBELAJARAN'] == model]

    prediksi_tuntas = data_model['Prediksi_Tuntas'].values[0]
    prediksi_tidak_tuntas = data_model['Prediksi_Tidak_Tuntas'].values[0]

    plt.figure(figsize=(6, 6))

    plt.pie(
        [prediksi_tuntas, prediksi_tidak_tuntas],
        labels=['Tuntas', 'Tidak Tuntas'],
        autopct='%1.1f%%',
        startangle=90
    )

    plt.title(f'Prediksi Ketuntasan Siswa pada Kelas {model}')
    plt.show()
```

Blok kode ini digunakan untuk membuat ringkasan *(summary)* hasil prediksi berdasarkan model pembelajaran, menghitung statistik penting, serta memvisualisasikan hasil prediksi ketuntasan siswa dalam bentuk diagram lingkaran *(pie chart)*. Tahap ini bertujuan untuk memudahkan analisis terhadap efektivitas setiap model pembelajaran berdasarkan hasil klasifikasi yang diperoleh dari model terbaik.

Pertama, program membuat DataFrame summary_model menggunakan fungsi groupby('MODEL_PEMBELAJARAN'). Fungsi ini digunakan untuk mengelompokkan data berdasarkan jenis model pembelajaran yang terdapat pada dataset. Setelah data dikelompokkan, fungsi agg() digunakan untuk menghitung beberapa statistik penting pada setiap kelompok model pembelajaran. Statistik yang dihitung meliputi jumlah siswa (Jumlah_Siswa) menggunakan fungsi count pada kolom NILAI, rata-rata nilai siswa (Rata_Rata_Nilai) menggunakan fungsi mean, jumlah siswa yang diprediksi Tuntas (Prediksi_Tuntas) menggunakan fungsi lambda untuk menghitung banyaknya data dengan nilai **"Tuntas"** pada kolom PREDIKSI_MODEL_TERBAIK, serta jumlah siswa yang diprediksi Tidak Tuntas (Prediksi_Tidak_Tuntas) dengan cara yang sama untuk kategori **"Tidak Tuntas"**. Setelah proses agregasi selesai, fungsi reset_index() digunakan agar hasil pengelompokan kembali menjadi DataFrame biasa dengan indeks yang teratur.

Berdasarkan summary model pembelajaran, diketahui bahwa jumlah siswa pada kelas CBL (Case-Based Learning) sebanyak 32 siswa dengan rata-rata nilai sebesar 85,29, sedangkan pada kelas PBL (Problem-Based Learning) terdapat 31 siswa dengan rata-rata nilai sebesar 78,36. Dari hasil prediksi menggunakan model Random Forest, sebanyak 25 siswa pada kelas CBL diprediksi mencapai ketuntasan dan 7 siswa diprediksi tidak tuntas. Sementara itu, pada kelas PBL terdapat 20 siswa yang diprediksi tuntas dan 11 siswa yang diprediksi tidak tuntas. Hasil ini menunjukkan bahwa kelas CBL memiliki jumlah siswa tuntas yang lebih banyak dibandingkan dengan kelas PBL.

Berdasarkan persentase prediksi ketuntasan, kelas CBL memiliki tingkat ketuntasan sebesar 78,13%, sedangkan persentase siswa yang diprediksi tidak tuntas sebesar 21,88%. Persentase tersebut mengindikasikan bahwa sebagian besar siswa pada kelas CBL diprediksi telah mencapai ketuntasan belajar. Selain itu, rata-rata nilai yang lebih tinggi pada kelas CBL dibandingkan dengan kelas PBL menunjukkan bahwa siswa pada kelas CBL secara umum memiliki performa akademik yang lebih baik.

Visualisasi diagram lingkaran (pie chart) semakin memperjelas distribusi ketuntasan siswa pada kelas CBL. Bagian berwarna biru yang mewakili kategori Tuntas mendominasi diagram dengan proporsi sebesar 78,1%, sedangkan bagian berwarna oranye yang mewakili kategori Tidak Tuntas hanya sebesar 21,9%. Hal ini menunjukkan bahwa lebih dari tiga perempat siswa pada kelas CBL berhasil mencapai ketuntasan belajar berdasarkan hasil prediksi model.

Secara keseluruhan, hasil ringkasan dan visualisasi tersebut menunjukkan bahwa model pembelajaran Case-Based Learning (CBL) memiliki rata-rata nilai yang lebih tinggi serta persentase ketuntasan yang lebih besar. Dengan demikian, berdasarkan data yang dianalisis menggunakan algoritma Random Forest, kelas CBL menunjukkan kecenderungan memiliki hasil belajar yang lebih baik dibandingkan dengan kelas PBL, yang ditunjukkan oleh dominasi jumlah siswa yang diprediksi mencapai ketuntasan belajar.

```python
# =========================
# KESIMPULAN
# =========================

model_pembelajaran_terbaik = summary_model.sort_values(
    by=['Persentase_Prediksi_Tuntas', 'Rata_Rata_Nilai'],
    ascending=False
).iloc[0]

print("\nKESIMPULAN:")
print("Model klasifikasi terbaik yang digunakan adalah Random Forest.")

print("Berdasarkan hasil prediksi Random Forest, model pembelajaran yang lebih baik adalah:",
      model_pembelajaran_terbaik['MODEL_PEMBELAJARAN'])

print("Persentase prediksi tuntas:",
      round(model_pembelajaran_terbaik['Persentase_Prediksi_Tuntas'], 2), "%")

print("Rata-rata nilai:",
      round(model_pembelajaran_terbaik['Rata_Rata_Nilai'], 2))
```

Blok kode ini digunakan untuk menentukan dan menampilkan kesimpulan akhir hasil analisis Data Mining berdasarkan hasil prediksi yang diperoleh dari model klasifikasi terbaik, yaitu Random Forest. Pada tahap ini, program melakukan identifikasi model pembelajaran yang memiliki performa terbaik berdasarkan persentase ketuntasan prediksi dan rata-rata nilai siswa.

Pertama, program membuat variabel model_pembelajaran_terbaik dengan memilih data terbaik dari DataFrame summary_model menggunakan fungsi sort_values(). Pengurutan dilakukan berdasarkan dua kriteria, yaitu Persentase_Prediksi_Tuntas dan Rata_Rata_Nilai, dengan parameter ascending=False yang berarti data diurutkan dari nilai terbesar ke nilai terkecil. Setelah data diurutkan, perintah .iloc[0] digunakan untuk mengambil baris pertama yang memiliki nilai tertinggi sehingga diperoleh model pembelajaran dengan performa terbaik berdasarkan hasil prediksi.

Selanjutnya, perintah print("\nKESIMPULAN:") digunakan untuk menampilkan judul KESIMPULAN pada output. Karakter \n berfungsi untuk membuat baris baru sehingga tampilan hasil menjadi lebih rapi dan mudah dibaca.

Pada baris berikutnya, program menampilkan informasi bahwa *Random Forest* merupakan model klasifikasi terbaik yang digunakan. Kesimpulan ini diperoleh dari tahap evaluasi model sebelumnya yang menunjukkan bahwa *Random Forest* memiliki performa paling baik dibandingkan algoritma lain yang diuji dalam penelitian.
